### Threshold Otsu

Em métodos padrões de Thresholding, utiliza-se um valor entre 0(preto) e 255(branco) para realizar a binarização
das imagens então por exemplo, se escolhermos o valor 127, pixels acima disso serão classificados como branco e abaixo disso como preto.

Em imagens bimodais,que possuem 2 picos em seu histograma, o que o algoritmo faz é escolher o melhor valor, aproximando-se do valor entre esses 2 picos. Então o algoritmo Otsu calcula automaticamente esse valor de threshold ótimo para imagens bimodais. 

![title](bimodal_image.png)

Exemplo de uma imagem bimodal, apos aplicado um filtro gaussiano, os 2 picos do seu histograma ficam claramente separados, propiciando o uso do algoritmo de Otsu.

Esse tipo de algoritmo é interessante de aplicar somente quando a imagem possui essa caracteristica bimodal, caso contrário, as demais técnicas de mean thresholding seriam mais aplicaveis.

```python
'''
Exemplode imagem bimodal com binarização pelo metodo Otsu
'''


%matplotlib inline

import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("noisy_leaf.jpg", 0)
plt.imshow(img, 'gray')
```

```python
'''
É possivel ver os 2 picos diferentes no histograma, indicando ser uma imagem bimodal
'''
plt.hist(img.ravel(), 256)
```

```python
'''
Imagem Binarizada
'''
ret, imgf = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(imgf,'gray')
```
